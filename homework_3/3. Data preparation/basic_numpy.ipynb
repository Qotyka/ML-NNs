##### 1. Создайте одномерный массив из чисел от 1 до 10. Затем преобразуйте его в двумерный массив 2x5. Найдите сумму всех элементов массива, а также среднее арифметическое.

```python
import numpy as np
# Одномерный массив
a = np.linspace(1, 10, 10)
print(a)
# Преобразуем в массив 2x5
b = a.reshape((2,5))
print(b)
# Находим сумму элементов
s = a.sum()
print(s)
# Находим среднее
mean = a.mean()
print(mean)
```

##### 2. Создайте матрицу размером 5x5, содержащую числа от 1 до 25. Извлеките из неё:
1. Первый столбец
2. Последние три элемента второй строки
3. Центральный элемент массива

```python
# Делаем матрицу 5x5
matrix = np.linspace(1, 25, 25)
matrix = matrix.reshape((5,5))
print(matrix)
# Извлекаем 1 столбец
a = matrix[:,0].reshape((5,1))
print(a)
# Последние 3 элемента 2 строки
b = matrix[1, 1:-1]
print(b)
# Центральный элемент (вариант 1)
c = matrix.reshape((1,25))[:, matrix.size//2]
print(c)
# Центральный элемент (вариант 2)
axis_size = int( (matrix.size)**0.5 )
c = matrix[axis_size//2, axis_size//2]
print(c)
```

##### 3. Создайте массив из 20 случайных чисел. Найдите:
1. Минимальное и максимальное значения
2. Медиану
3. Стандартное отклонение

```python
# Одномерный массив случайных чисел
rarr = np.random.randint(1, 100, 20)
print(rarr)
# Минимальное число
print(rarr.min())
# Максимальное число
print(rarr.max())
# Медиана
print(np.median(rarr))
# Стандартное отклонение
print(np.std(rarr))
```

##### 4. Создайте массив случайных чисел от 0 до 100 (10 элементов). Определите, какие из них чётные. Замените все нечётные элементы на 0.

```python
rarr = np.random.randint(0, 100, 10)
print(rarr)
# Создаем массив-маску (определяем нечетные элементы)
mask = rarr % 2 != 0
# Приравниваем все элементы, удовлетворяющие маске нулю
rarr[mask] = 0
print(rarr)
```

##### 5.1 Создайте матрицу произвольного размера. 
1. Найдите транспонированную матрицу.
2. Найдите обратную матрицу, если это возможно.
3. Найдите собственные значения и собственные вектора.

```python
axis_size = 3
arr = np.random.randint(0, 100, axis_size**2)
arr = arr.reshape((axis_size, axis_size))
print(arr)
# Транспонированная
print(arr.T)
# Обратная
try:
    inv = np.linalg.inv(arr)
except:
    print("impossible to inv the matrix")
print(inv)
# Собственные значения и собственные вектора
lambd, vectors = np.linalg.eig(arr)
print(lambd)
print(vectors)
```

##### 5.2 Создайте две матрицы 4x4, заполненные случайными числами. Выполните следующие операции:

1. Найдите их матричное произведение.
2. Объедините две матрицы в тензор/в одну матрицу 4x8.

```python
a = np.random.randint(1, 100, 16).reshape((4, 4))
b = np.random.randint(1, 100, 16).reshape((4, 4))
print(a)
print(b)
# Произведение
print(a.dot(b))
# 4x8 матрица - объединение по последнему индексу (оси)
c = np.hstack((a, b))
print(c)
# тензор
c = np.stack((a,b))
print(c)
```

##### 6. Создайте трёхмерный массив размером 3x3x3, заполненный случайными числами. Выполните следующие операции:
1. Найдите сумму элементов по каждой оси.
2. Найдите минимальный элемент для каждого слоя.
3. Отсортируйте каждый слой по возрастанию.

```python
a = np.random.randint(1, 100, 27).reshape((3, 3, 3))
print(a)
# Сумма элементов по осям
for i in range(3):
    print(a.sum(axis= i))
# Минимальный элемент слоя
print(a.min(axis=2).min(axis= 1))
# Сортируем каждый слой
for j in range(3):
    a[j, :, :] = np.sort(a[j, :, :], axis= None).reshape((3, 3))
print(a)
```

##### 7. Сгенерируйте набор данных из 100 случайных точек в 3D-пространстве. Выполните их нормализацию (приведите к единичному масштабу) по каждой оси.

```python
a = np.float64(np.random.randint(1, 100, 3*100).reshape((100, 3)))
for i in range(3):
    a[:, i] = a[:, i] / a[:, i].max()
print(a)
```

##### 8*. Загрузите любое изображение (в рамках приличия, конечно) и сожмите его с помощью сингулярного разложения (SVD). Выберете разное количество сингулярных значений и сравните полученный результат с оригиналом.

```python
from PIL import Image
import matplotlib.pyplot as plt
im = Image.open("4x_BW.webp").convert(mode= "L")
im_arr = np.array(im)
U, S, Vh = np.linalg.svd(im_arr)
# Разложили и теперь мы можем хранить не m*n значений (16.384), а
# r*(m + n + 1) = 7.710 (r = 30)
for r in range(10, 100, 10):
    new_im_arr = U[:, :r].dot(np.diag(S)[:r, :r].dot(Vh[:r, :])) 
    new_im = Image.fromarray(new_im_arr)
    plt.subplot(3, 3, r//10)
    plt.imshow(new_im)
    plt.title(f"r is {r}")
```
